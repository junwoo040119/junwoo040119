```python
import torch
import torchvision.models as models
```

```python
# 학습된 모델 생성 및 가중치 저장
model = models.vgg16(weights='IMAGENET1K_V1')
torch.save(model.state_dict(), 'mode_weights.pth')
```

```python
# 모델 불러오기
model = models.vgg16() # 학습되지 않은 모델 생성
model.load_state_dict(torch.load('mode_weights.pth'))
model.eval()  # 평가 모드로 전환
```

```python
# 모델 형태를 포함하여 저장 및 불러오기
torch.save(model, 'complete_model.pth') # 모델 형태 포함 저장하기
model = torch.load('complete_model.pth') # 모델 불러오기
```
